```python
import pandas as pd
```

```python
cpc = pd.read_csv('all/cpc2_cc_codpot_all.txt.txt', sep="\t")
```

```python
cpc
```

```python
# filter for the label 'nondocing' and a transcript_length equal to or longer than 200
cpc = cpc[(cpc['label'] == 'noncoding') & (cpc['transcript_length'] >= 200)]

# convert everything in lower case for the ID column 
cpc['#ID'] = cpc['#ID'].str.lower()
```

```python
cpc
```

```python
# export the filtered data to a new file
cpc.to_csv('cpc2_lncRNA.csv', sep="\t", index=False)
```
