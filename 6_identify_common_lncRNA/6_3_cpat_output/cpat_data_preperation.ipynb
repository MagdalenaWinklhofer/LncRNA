# Extract gene id

```python
# check the ID name beginnings.
# I want to know how many different versions of the transcripts are present in the file.

import csv

filename = '/home/maggy/LncRNA/5_4_cpat/cpat_lncRNA_sense_topORF/cpat_lncRNA_sense_ORF1.ORF_prob.tsv'
versions = set()

with open(filename, 'r') as file:
    reader = csv.reader(file, delimiter='\t')
    next(reader)  # Skip the header row
    for row in reader:
        name = row[0]
        version = name[:5] # Extract the first 5 characters
        versions.add(version)

number_of_versions = len(versions)
print("There are", number_of_versions, "different versions:", versions)
```

```python
import pandas as pd 

cpat=pd.read_csv('/home/maggy/LncRNA/5_4_cpat/cpat_lncRNA_sense_topORF/cpat_lncRNA_sense_ORF1.ORF_prob.tsv', sep='\t')

for i in range(len(cpat)):
    name = cpat["ID"][i].lower()
    #name = name[:-6] # cut the _ORF_numer from the back of the name
    if name.startswith("strg"):
        lines = name.split("_")
        line = lines[0]
        cpat.loc[i, 'transcript_id'] = line
    else:
        print("Row not in entries:")
        print(name)
        break
```

# Filter for non-coding 

```python
# filter df_cpat for coding probability > 0.364 (for zebrafish it is 0.381 (see cutoff.txt))
# I only need to keep the transcripts that have a coding probability lower than 0.364
cpat = cpat[cpat['Coding_prob'] <= 0.381]
cpat = cpat[cpat['mRNA'] >= 200]
```

```python
# print out how many transcript start with scaff, mstrg, ccar
print('scaff:',cpat['transcript_id'].str.startswith('scaff').sum())
print('mstrg:', cpat['transcript_id'].str.startswith('mstrg').sum())
print('ccar:', cpat['transcript_id'].str.startswith('ccar').sum())
print('strg:', cpat['transcript_id'].str.startswith('strg').sum())
```
